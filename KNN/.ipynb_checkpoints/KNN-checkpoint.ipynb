# Data Preprocessing

```python
df.select_dtypes(include='object').head()
```

```python
encoder = LabelEncoder()
```

```python
df['income'] = encoder.fit_transform(df['income'])
```

```python
df['income'].value_counts()
```

```python
sns.countplot(x=df['income'], data=df, palette='Set2')
plt.xlabel('Count of <=50K & >50K')
plt.ylabel('Count')
plt.title('Distribution of <=50K & >50K')
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2,  # x-coordinate
             p.get_height() + 10,            # y-coordinate
             int(p.get_height()),           # The count value
             ha='center')                   # Center alignment
plt.xticks([0, 1], ['<=50K', '>50K'])
plt.show()
```

```python
x = df.drop('income', axis=1)
x.head()
```

```python
y = df['income']
y.head()
```

```python
x.nunique()
```

```python
X = pd.get_dummies(x, dtype='int', drop_first=True)
X.head()
```

```python
X.shape
```

```python
y.shape
```

# Split dataset into training and testing sets

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

# Scaling

```python
scaler = StandardScaler()
```

```python
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)
```

# Machine Learning Models

```python
model = KNeighborsClassifier()
```

```python
model.fit(xtrain_scaled, ytrain)
```

```python
y_pred_knn = model.predict(xtest_scaled)
y_pred_knn
```

```python
model.score(xtrain_scaled, ytrain)
```

```python
model.score(xtest_scaled, ytest)
```

```python
accuracy_score(y_pred_knn, ytest)
```

```python
cm = confusion_matrix(y_pred_knn, ytest)
cm
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
```

```python
print(classification_report(y_pred_knn, ytest))
```

```python
RocCurveDisplay.from_predictions(y_pred_knn, ytest)
plt.plot([0,1], [0,1])
plt.title('Receiver Operating Characteristic (ROC)')
plt.show()
```
