```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('logistic_regression_dataset.csv')
```

```python
df.head()
```

```python
plt.scatter(df.Age,df.Purchased,marker = '+',color = 'red')
```

```python
plt.scatter(df.Income,df.Purchased,marker = '+',color = 'red')
```

```python
plt.scatter(df.Experience,df.Purchased,marker = '+',color = 'red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['Age','Income','Experience']]
y = df['Purchased']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)
```

```python
X_test.info()
```

```python
X_train.info()
```

```python
X_train = X_train.reset_index(drop=True)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear')
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
# Get predictions
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
